```python
import csv

data = None

with open('/content/squads.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    data = [row for row in csv_reader]

data
```

#ASSIGNMENT 1

```python
players_list = []

for elem in data:
  players_list.append({
    'number': data[data.index(elem)]["1"],
    'position': data[data.index(elem)]["GK"],
    'name': data[data.index(elem)]["Ángel Bossio"],
    'date_of_birth': data[data.index(elem)]["(1905-05-05)5 May 1905 (aged 25)"],
    'caps': data[data.index(elem)][""],
    'club': data[data.index(elem)]["Talleres"],
    'country': data[data.index(elem)]["Argentina"],
    'club_country': data[data.index(elem)]["Argentina"],
    'year': data[data.index(elem)]["1930"]
  })

print(players_list[:5])
```

#ASSIGNMENT 2

```python
position_dict = {
    "GK": [],
    "DF": [],
    "MF": [],
    "FW": [],
}

for elem in players_list:
    if elem["position"] == "GK":
      position_dict["GK"].append(elem)
    elif elem["position"] == "DF":
      position_dict["DF"].append(elem)
    elif elem["position"] == "MF":
      position_dict["MF"].append(elem)
    elif elem["position"] == "FW":
      position_dict["FW"].append(elem)

position_dict
```

#ASSIGNMENT 3

```python
players_country ={
    "Argentina": {},
    "Brazil": {}
}

temp_dict = {
    "GK": [],
    "DF": [],
    "MF": [],
    "FW": [],
   }
temp2_dict = {
    "GK": [],
    "DF": [],
    "MF": [],
    "FW": [],
   }
for elem in players_list:

  if elem["country"] == "Argentina":
          if elem["position"] == "GK":
            temp_dict["GK"].append(elem)
          elif elem["position"] == "DF":
            temp_dict["DF"].append(elem)
          elif elem["position"] == "MF":
            temp_dict["MF"].append(elem)
          elif elem["position"] == "FW":
            temp_dict["FW"].append(elem)
  elif elem["country"] == "Brazil":
          if elem["position"] == "GK":
            temp2_dict["GK"].append(elem)
          elif elem["position"] == "DF":
            temp2_dict["DF"].append(elem)
          elif elem["position"] == "MF":
            temp2_dict["MF"].append(elem)
          elif elem["position"] == "FW":
            temp2_dict["FW"].append(elem)

players_country["Brazil"].update(temp2_dict)
players_country["Argentina"].update(temp_dict)
```
